```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from  mlxtend.preprocessing import TransactionEncoder
```

```python
df = pd.read_csv('Global_YouTube_Statistics.csv',encoding='iso-8859-1')
```

```python
df_cleaned = df.replace('nan', pd.NA)  # Replace 'nan' with pandas' NA for numeric columns
numeric_columns = ['rank', 'subscribers', 'video views', 'video_views_for_the_last_30_days',
                   'lowest_monthly_earnings', 'highest_monthly_earnings', 'lowest_yearly_earnings',
                   'highest_yearly_earnings', 'subscribers_for_last_30_days', 'created_year', 'Unemployment rate']

for col in numeric_columns:
    df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')
    
df_cleaned = df_cleaned.dropna()
```

```python
df_cleaned
```

```python
df_cleaned.reset_index(inplace = True,drop=True)
df_cleaned
```

```python
country = []
for i in range(0,df_cleaned['Country'].shape[0]):
    country.append([df_cleaned['Country'].iloc[i]])
```

```python
country
```

```python
te = TransactionEncoder() 
te_ary = te.fit_transform(country)
new_country = pd.DataFrame(te_ary,columns=te.columns_)
new_country
```

```python
att = pd.concat([df_cleaned['subscribers'],df_cleaned['video_views_for_the_last_30_days'],new_country],axis=1)
label = df_cleaned['lowest_monthly_earnings']
```

```python
att
```

```python
label
```

```python
model = LinearRegression()
model.fit(att,label)
model.score(att,label)
```

```python
pf = PolynomialFeatures(degree= 2)
pf.fit(att)
att_new = pf.transform(att)

model = LinearRegression()
model.fit(att_new,label)
model.score(att_new,label)
```
